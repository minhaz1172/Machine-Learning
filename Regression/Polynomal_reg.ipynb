# Apply Polynomial Regression on a real estate dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures
```

# Load the dataset

```python
df=pd.read_csv('C:\AI and ML\Machine Learning\Machine-Learning\Regression\Regression\Real estate.csv')
df.head()
```

# dEFINE X AND y

```python
X=df[['X2 house age','X5 latitude','X6 longitude']]
y=df['Y house price of unit area']
```

```python
#Apply Polynomial Featutes
poly= PolynomialFeatures(degree=2, include_bias=False)
X_poly=poly.fit_transform(X)
```

# Split the data

```python
X_train, X_test, y_train, y_test = train_test_split(X_poly, y, test_size=0.2, random_state=42)
```

# Train The Model

```python
model= LinearRegression()
model.fit(X_train, y_train)
```

```python
# predictions and evaluation
y_pred=model.predict(X_test)
mse= mean_squared_error(y_test, y_pred)
r2=r2_score(y_test, y_pred)
print(f"mean_squared_error: {mse}")
print(f"r2_score: {r2}")
```

```python
# Plot the results
plt.Figure(figsize=(10, 6))
#plt.subplot(1, 2, 1)
plt.scatter(df['X2 house age'],y, color='blue', label='')
plt.xlabel('X2 house age')
plt.ylabel('Y house price of unit area')
plt.title('Polynomial Regression')
plt.grid(True)
plt.show()
```
